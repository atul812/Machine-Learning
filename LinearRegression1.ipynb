```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model

value = pd.read_csv("homeprices.csv")
value

plt.scatter(value.area, value.price, color='red', marker='.')
plt.xlabel('sq ft (Area)')
plt.ylabel('prices')
```

```python
reg = linear_model.LinearRegression()
reg.fit(value[['area']], value.price)
```

```python
reg.predict([[3456]])
```

```python
reg.coef_ #coefficient=m (y=mx+b)
```

```python
reg.intercept_ #y intercept=b (y=mx+b)
```

```python
new_data = pd.read_csv('areas.csv')
new_data
```

```python
p = reg.predict(new_data)
new_data['prices'] = p
new_data
```

```python
plt.scatter(value.area, value.price, color='red')
plt.plot(value.area, reg.predict(value[['area']]), color='blue')
```

```python
# to put the trained model as a file

import pickle
with open('model_pickle', 'wb') as f:
    pickle.dump(reg, f)
```

```python
#to use the file which stored the model

with open('model_pickle', 'rb') as f:
    model = pickle.load(f)
model.predict([[2349]])
```
