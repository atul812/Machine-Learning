```python
# hiring.csv This file contains hiring statics for a firm such as experience of candidate, his written test score and personal interview score. Based on these 3 factors, HR will decide the salary. Given this data, you need to build a machine learning model for HR department that can help them decide salaries for future candidates. 
# Using this predict salaries for following candidates
# 2 yr experience, 9 test score, 6 interview score
# 12 yr experience, 10 test score, 10 interview score


import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model

val = pd.read_csv('hiring.csv')
val
```

```python
#preprocess the data

from word2number import w2n
val['experience'] = val['experience'].fillna('zero')
val['experience']
```

```python
#in column use .apply(w2n.word_to_num)
#val['experience'] = val['experience'].apply(w2n.word_to_num)
#val
```

```python
import math
med = val.experience.median()
for i in range(len(val)): 
    if val.experience[i]==0:
        val.experience[i] = med
val.experience
```

```python
median = val['test_score(out of 10)'].median()
val['test_score(out of 10)'] = val['test_score(out of 10)'].fillna(median)
val['test_score(out of 10)']
```

```python
val
```

```python
# train the model

reg = linear_model.LinearRegression()
reg.fit(val[['experience', 'test_score(out of 10)', 'interview_score(out of 10)']], val['salary($)'])
```

```python
reg.predict([[2, 9, 6]])
```

```python
reg.coef_
```
