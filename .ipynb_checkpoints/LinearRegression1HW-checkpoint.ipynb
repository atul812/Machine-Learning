```python
# predict the per capita income of year 2020

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model

val = pd.read_csv('canada_per_capita_income.csv')
pci = val['per capita income (US$)']
plt.plot(val.year, pci)
```

```python
regr = linear_model.LinearRegression()
x = val[['year']]
y = pci
regr.fit(x, y)
```

```python
regr.predict([[2020]])
```
