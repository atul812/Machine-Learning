```python
import numpy as np
import pandas as pd

data = pd.read_csv('homeprices3.csv')
data
```

```python
dummy = pd.get_dummies(data.town)
dummy
```

```python
merge = pd.concat([data, dummy], axis='columns')
merge
```

```python
final = merge.drop(['town', 'west windsor'], axis='columns')
final
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
x = final.drop(['price'], axis='columns')
x
```

```python
y = final.price
y
```

```python
model.fit(x, y)
```

```python
model.predict([[3123, 0, 1]])
```

```python
model.predict([[3233, 0, 0]])
```

```python
#to check the model score
model.score(x, y)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
dfle = data
dfle.town = le.fit_transform(dfle.town)
dfle
```

```python
X = data[['town', 'area']].values
X
```

```python
y = data['price']
y
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
ct = ColumnTransformer(
    [("encoder", OneHotEncoder(), [0])],
    remainder="passthrough"
)
X = ct.fit_transform(X)
X
```

```python
from sklearn.pipeline import Pipeline
pipeline = Pipeline([
    ("preprocessing", ct),
    ("model", LinearRegression())
])

# Train
pipeline.fit(X, y)

# Predict 
pipeline.predict([[1, 2800]])
```
